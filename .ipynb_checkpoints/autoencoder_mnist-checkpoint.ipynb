# Convolutional Autoencoder - MNIST

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
import torch
from torchvision import datasets,transforms
from torch.utils.data import DataLoader,SubsetRandomSampler,sampler
import time
import numpy as np
import matplotlib.colors as mcolors
```

```python
## check available device
device = (torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu'))
print("device:", device)
```

```python
batch_size = 32
lr = 0.0005
n_epochs = 40
```

```python
train_dataset = datasets.MNIST(root='data', train=True, transform=transforms.ToTensor(), download=True)
len(train_dataset)
```

```python
train_dataset
```

```python
## select a smaller set to test
train_indices = torch.arange(0, 1000)
#print(train_indices)

train_sampler = SubsetRandomSampler(train_indices)

#train_loader = DataLoader(dataset=train_dataset, batch_size=batch_size, num_workers=0,shuffle=True)
train_loader = DataLoader(dataset=train_dataset, batch_size=batch_size, num_workers=0,sampler=train_sampler)
```

```python
# Check the dataset
for images, labels in train_loader:  
    print('Image batch dimensions:', images.size())
    print('Image label dimensions:', labels.size())
    print(labels[:10])
    break
```

```python
class Reshape(nn.Module):
    def __init__(self, *args):
        super().__init__()
        self.shape = args

    def forward(self, x):
        return x.view(self.shape)
    
class Trim(nn.Module):
    def __init__(self, *args):
        super().__init__()

    def forward(self, x):
        return x[:, :, :28, :28]
```

## Model

```python
class AutoEncoder(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.encoder = nn.Sequential(
                nn.Conv2d(1, 32, stride=(1, 1), kernel_size=(3, 3), padding=1),
                nn.LeakyReLU(0.01),
                nn.Conv2d(32, 64, stride=(2, 2), kernel_size=(3, 3), padding=1),
                nn.LeakyReLU(0.01),
                nn.Conv2d(64, 64, stride=(2, 2), kernel_size=(3, 3), padding=1),
                nn.LeakyReLU(0.01),
                nn.Conv2d(64, 64, stride=(1, 1), kernel_size=(3, 3), padding=1),
                nn.Flatten(),
                nn.Linear(3136, 2)
        )
        self.decoder = nn.Sequential(
                nn.Linear(2, 3136),
                Reshape(-1, 64, 7, 7),
                nn.ConvTranspose2d(64, 64, stride=(1, 1), kernel_size=(3, 3), padding=1),
                nn.LeakyReLU(0.01),
                nn.ConvTranspose2d(64, 64, stride=(2, 2), kernel_size=(3, 3), padding=1),                
                nn.LeakyReLU(0.01),
                nn.ConvTranspose2d(64, 32, stride=(2, 2), kernel_size=(3, 3), padding=0),                
                nn.LeakyReLU(0.01),
                nn.ConvTranspose2d(32, 1, stride=(1, 1), kernel_size=(3, 3), padding=0), 
                Trim(),  # 1x29x29 -> 1x28x28
                nn.Sigmoid()
                )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

```python
model = AutoEncoder()
model.to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=lr)  
```

## Training

```python
def train_epoch(model,n_epochs, optimizer, train_loader):
    
    loss_fn = nn.MSELoss()
    train_loss = []
    
    for epoch in range(n_epochs):
        model.train()

        loss_all = 0
        for batch_idx, (features, _) in enumerate(train_loader):

            features = features.to(device)
            preds = model(features)
            loss = loss_fn(preds, features)
            optimizer.zero_grad()
            loss.backward()
            # update model parameters:
            optimizer.step()
            
            loss_all += loss.detach().cpu().numpy()
        loss_all /= len(features)
        train_loss.append(loss_all)

    return train_loss
```

```python
train_loss = train_epoch(model, n_epochs, optimizer=optimizer, train_loader=train_loader)

print(train_loss)

plt.plot(train_loss,"o--")
```

## Evaluation

```python
n_imgs = 15
for batch_idx, (features, _) in enumerate(train_loader):
    features = features.to(device)
    orig_images = features[:n_imgs]

    with torch.no_grad():
        decoded_images = model(orig_images)
        
    break
```

```python
## plot
fig, axes = plt.subplots(nrows=2, ncols=n_imgs, sharex=True, sharey=True, figsize=(20,2.5))

for i in range(n_imgs):
    for ax, img in zip(axes, [orig_images, decoded_images]):
        curr_img = img[i].detach().to(torch.device('cpu'))
        ax[i].imshow(curr_img[0], cmap='binary')
```

```python
n_classes = 10

d = {i:[] for i in range(n_classes)}

model.eval()
with torch.no_grad():
    for i, (features, targets) in enumerate(train_loader):

        features = features.to(device)
        targets = targets.to(device)

        embedding = model.encoder(features)

        for i in range(n_classes):
            if i in targets:
                mask = targets == i
                d[i].append(embedding[mask].to('cpu').numpy())

colors = list(mcolors.TABLEAU_COLORS.items())
print("colors:", colors)
for i in range(n_classes):
    d[i] = np.concatenate(d[i])
    plt.scatter(d[i][:, 0], d[i][:, 1], color=colors[i][1], label=f'{i}', alpha=0.5)

plt.legend()
```
